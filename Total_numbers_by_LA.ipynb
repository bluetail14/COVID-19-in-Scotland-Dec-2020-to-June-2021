## Cases and Deaths

```python
df = pd.read_csv('../MyVizProject/data_cases_la.csv',na_values='',encoding = "ISO-8859-1")
df.head()
```

```python
df = df.rename(columns = {'ï»¿_id':'id'})
df["Date"] = pd.to_datetime(df.loc[:, 'Date'], format='%Y%m%d')
```

```python
df_cases = df[(df.CAName!='Unknown')]
df_cases = df_cases[[ 'Date', 'CAName', 'DailyPositive']]#.set_index('Date')
df_cases.head()
```

```python
sc = df_cases.groupby(['Date', 'CAName'])['DailyPositive'].sum()
sc
```

```python
df_cases_g = sc.unstack(level=1)
df_cases_g.tail(10)
```

```python
df_c = ((df_cases_g.rolling(7).mean())).shift(-7).rolling(1).mean().fillna(0)
# shift(-7) moves it back by 7 days to the day of exposure.
df_c.tail(10)
```

```python
df_c = df_c.stack().copy()
df_c
```

```python
df_c1 = df_c.to_frame().reset_index()
df_c1.set_index('Date',inplace=True)
df_c1.rename(columns={'CAName':'Council',0:'Positive_Cases_per_LA'}, inplace=True)
df_c1.Positive_Cases_per_LA = df_c1.Positive_Cases_per_LA.round(0).astype(int)
df_c1
```

```python
cases = df_c1.loc[(df_c1.index>='2021-01-05')&(df_c1.index<='2021-06-30')]
cases.to_csv('output_cases_la.csv')
#cases.tail(20)
```

```python
df_deaths = df[(df.CAName!='Unknown')]
df_deaths = df_deaths[[ 'Date', 'CAName', 'DailyDeaths']]#.set_index('Date')
df_deaths.head()
```

```python
sd = df_deaths.groupby(['Date', 'CAName'])['DailyDeaths'].sum()
sd
```

```python
df_deaths_g = sd.unstack(level=1)
df_deaths_g.tail(10)
```

```python
df_d = (df_deaths_g.rolling(7).mean()).shift(-14).rolling(1).mean().fillna(0)
df_d.tail(30)
```

```python
df_d = df_d.stack().copy()
df_d
```

```python
df_d1 = df_d.to_frame().reset_index()
df_d1.set_index('Date',inplace=True)
df_d1.rename(columns={'CAName':'Council',0:'Deaths_per_LA'}, inplace=True)
df_d1.Deaths_per_LA = df_d1.Deaths_per_LA.round(0).astype(int)
df_d1.tail(30)
```

```python
deaths = df_d1.loc[(df_d1.index>='2021-01-05')&(df_d1.index<='2021-06-30')]
deaths.to_csv('output_deaths_la.csv')
#deaths.tail()
```
