```python
import pandas as pd
import numpy as np
```

```python
dfv = pd.read_csv('../MyVizProject/data_vacc_age.csv', na_values='',  parse_dates=True, encoding = "ISO-8859-1")
dfv.head()
```

```python
df_v = dfv[['Date', 'Sex','AgeGroup','NumberVaccinated']]
df_v["Date"] = pd.to_datetime(df_v.loc[:, 'Date'], format='%Y%m%d')
```

```python
dfv.AgeGroup.unique()
```

```python
dfv.AgeGroup.unique()
```

```python
df_v = df_v[(df_v.Sex =='Total')]
df_v = df_v[['Date','AgeGroup', 'NumberVaccinated']]
df_v.head()
```

```python
df_v = df_v.groupby(['Date', 'AgeGroup'])['NumberVaccinated'].sum().reset_index()
df_v
```

```python
df_va = df_v.loc[(df_v.AgeGroup=='18 - 29')|(df_v.AgeGroup=='30 - 39')|(df_v.AgeGroup=='40 - 49')|(df_v.AgeGroup=='50 - 54')|(df_v.AgeGroup=='55 - 59')|(df_v.AgeGroup=='60 - 64')|(df_v.AgeGroup=='65 - 69')|(df_v.AgeGroup=='70 - 74')|(df_v.AgeGroup=='75 - 79')|(df_v.AgeGroup=='80 years and over')]
df_va.head(11)
```

```python
df_v.AgeGroup.unique()
```

```python
m = df_va['AgeGroup'].str.extract('(\d+)', expand=False).astype(int) < 60
```

```python
df_va['AgeGroup'] = np.where(m, 'under 60', 'over 60')
df_va = df_va.groupby(['Date','AgeGroup']).sum().reset_index()
df_va.set_index('Date', inplace = True)
df_va
```

```python
df_v75 = df_v.loc[(df_v.AgeGroup=='75 - 79')|(df_v.AgeGroup=='80 years and over')]
df_v75 = df_v75.groupby(['Date']).sum()
df_v75.insert(1, 'AgeGroup', 'over 75')
df_v75 = df_v75[['AgeGroup','NumberVaccinated']]
df_v75
```

```python
df_mva = df_va.append(df_v75).sort_values(by='Date').reset_index()
df_mva.head()
```

```python
df_mva.tail(20)
```

```python
df_vaccs_over60 = (((df_mva_over60.rolling(7)).mean())).shift(22).rolling(68, min_periods=1).sum().fillna(0).round()
df_vaccs_over60.NumberVaccinated = df_vaccs_over60.NumberVaccinated.astype(int)
df_vaccs_over60['AgeGroup'] = 'over 60'
df_vaccs_over60.tail()
```

```python
df_mcd.head()
```

```python
df_mcd_under60 = df_mcd.loc[df_mcd.AgeGroup=='under 60']
df_mcd_under60.head()
```

```python
df_mcd_over60 = df_mcd.loc[df_mcd.AgeGroup=='over 60']
df_mcd_over60.head()
```

```python
df_mcd_over75 = df_mcd.loc[df_mcd.AgeGroup=='over 75']
df_mcd_over75.head()
```

```python
df_mcd_under60.DailyPositive = ((df_mcd_under60.DailyPositive.rolling(7).mean())).shift(-7).rolling(1).mean().fillna(0)
df_mcd_under60.tail(12)
```

```python
df_mcd_over60.DailyPositive = ((df_mcd_over60.DailyPositive.rolling(7).mean())).shift(-7).rolling(1).mean().fillna(0)
df_mcd_over60.head()
```

```python
df_mcd_over75.DailyPositive = ((df_mcd_over75.DailyPositive.rolling(7).mean())).shift(-7).rolling(1).mean().fillna(0)
df_mcd_over75.head()
```

```python
n = (df_mcd_under60.DailyDeaths.rolling(7).mean())
#p = (df_mcd_under60.DailyDeaths.shift(7).rolling(7).mean())
#noinf = n/p
#n.replace(np.inf, 0.0000001, inplace=True)

no_inf = (n).shift(-14).rolling(1).mean().fillna(0)

no_inf.tail(12)#

#noinf[noinf == np.inf]
```

```python
df_mcd_under60.DailyDeaths = no_inf
df_mcd_under60.head(12)
```

```python
n1 = (df_mcd_over60.DailyDeaths.rolling(7).mean())
#p = (df_mcd_over60.DailyDeaths.shift(7).rolling(7).mean())
#noinf = n/p
#noinf.replace(np.inf, 0.0000001, inplace=True)

no_inf1 = (n1).shift(-14).rolling(1).mean().fillna(0)

no_inf1.tail()
```

```python
df_mcd_over60.DailyDeaths = no_inf1
df_mcd_over60.head()
```

```python
n2 = (df_mcd_over75.DailyDeaths.rolling(7).mean())
#p = (df_mcd_over75.DailyDeaths.shift(7).rolling(7).mean())
#noinf = n/p
#noinf.replace(np.inf, 0.0000001, inplace=True)

no_inf2 = (n2).shift(-14).rolling(1).mean().fillna(0)

no_inf2.tail()
```

```python
df_mcd_over75.DailyDeaths = no_inf2
df_mcd_over75.head()
```

```python
df_cd_adj0 = df_mcd_under60.append(df_mcd_over60).sort_values(by='Date').reset_index()
df_cd_adj = df_cd_adj0.append(df_mcd_over75).sort_values(by='Date').reset_index()
df_cd_adj.drop(columns = {'index','level_0'}, axis=1, inplace=True)
df_cd_adj.head(40)
```

```python
df_ages_rel0 = df_cd_adj.merge(df_va_adj, how='inner' , on=["Date", "AgeGroup"]).sort_values(by='Date').reset_index()
#df_ages_growth.set_index('Date', inplace=True)
#df_ages_growth.reset_index(inplace=True)
df_ages_rel0.drop('index', axis=1, inplace=True)
df_ages_rel = df_ages_rel0.round(decimals=0)
df_ages_rel.DailyPositive = df_ages_rel.DailyPositive.astype(int)
df_ages_rel.DailyDeaths = df_ages_rel.DailyDeaths.astype(int)
df_ages_rel.NumberVaccinated = df_ages_rel.NumberVaccinated.astype(int)
df_ages_rel.head(32)
```
